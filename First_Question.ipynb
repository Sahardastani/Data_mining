```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
xz=new_df_zscore.iloc[:,2:8]
xz
```

```python
yz=new_df_zscore.iloc[:,-1]
yz
```

```python
xz_train,xz_test,yz_train,yz_test=train_test_split(xz,yz,test_size=0.3,random_state=101)
```

```python
lm=LinearRegression()
lm.fit(xz_train,yz_train)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
coef_df=pd.DataFrame(lm.coef_,xz.columns,columns=['Coefficient'])
coef_df
```

```python
prediction_test=lm.predict(xz_test)
plt.scatter(yz_test,prediction_test)
```

```python
prediction_train=lm.predict(xz_train)
plt.scatter(yz_train,prediction_train)
```

```python
print('MAE:', metrics.mean_absolute_error(yz_test,prediction_test))
print('MSE:', metrics.mean_squared_error(yz_test,prediction_test))
print('RMSE:', np.sqrt(metrics.mean_squared_error(yz_test,prediction_test)))
```

```python
print('MAE:', metrics.mean_absolute_error(yz_train,prediction_train))
print('MSE:', metrics.mean_squared_error(yz_train,prediction_train))
print('RMSE:', np.sqrt(metrics.mean_squared_error(yz_train,prediction_train)))
```

```python
result=pd.DataFrame({'Actual':yz_test,'Predicted':prediction_test})
result
```
